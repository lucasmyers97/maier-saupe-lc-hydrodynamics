```python
from sympy import *
```

## Two disclination configuration is given by:
\begin{equation}
    \theta_\text{iso}
    =
    \frac12 \left[
        \text{atan2}\left(y, x + \frac{d}{2} \right)
        + \text{atan2} \left(y, x - \frac{d}{2}\right)
    \right]
\end{equation}

## Confinement boundary condition is given by:
\begin{equation}
    \left. \theta \right|_{\partial \Omega}
    =
    \varphi
    =
    \text{atan2} (y, x)
\end{equation}

## Hence, the boundary Poisson equation values must add to:
\begin{equation}
    \theta_c
    =
    \text{atan2} (y, x)
    - \frac12 \left[
        \text{atan2}\left(y, x + \frac{d}{2} \right)
        + \text{atan2} \left(y, x - \frac{d}{2}\right)
    \right]
\end{equation}

## Let's do a Taylor Series expansion about $d / 2$

```python
x, y = symbols('x y', real=True)
d = symbols('d', positive=True)
```

```python
theta = atan2(y, x) - Rational(1, 2) * (atan2(y, x + d/2) + atan2(y, x - d/2))

display(theta)
```

```python
series(theta, d, 0, 20)
```
